# Load EBM data

```python
import pandas as pd
import numpy as np


file_path = 'D:/VScode/EBM_NH_SH_sea_ice/EBM_data/'


# Replace these paths with the actual paths to your CSV files
nh_csv_file_path = file_path+'results_nh_n400_age0to200ka_20240208_0853.csv'
sh_csv_file_path = file_path+'results_sh_n400_age0to200ka_20240208_0853.csv'

# Read the CSV files
results_nh_df = pd.read_csv(nh_csv_file_path)
results_sh_df = pd.read_csv(sh_csv_file_path)

# rename columns to 'age' 'T' and 'iceA'
results_nh_df.columns = ['age', 'T', 'iceA']
results_sh_df.columns = ['age', 'T', 'iceA']
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 1, figsize=(10, 7), dpi=600, sharex=True)

# Northern Hemisphere Plot
ax1 = axs[0]
ax2 = ax1.twinx()  # Create a second y-axis for the iceA data
ax1.plot(results_nh_df['age'], results_nh_df['T'], 'r-', label='60°N Surface Air Temperature')
ax2.plot(results_nh_df['age'], results_nh_df['iceA'], 'b-', label='NH Ice Area')

# Southern Hemisphere Plot
ax3 = axs[1]
ax4 = ax3.twinx()  # Create a second y-axis for the iceA data
ax3.plot(results_sh_df['age'], results_sh_df['T'], 'r-', label='60°S Surface Air Temperature')
ax4.plot(results_sh_df['age'], results_sh_df['iceA'], 'b-', label='SH Ice Area')

# Titles and labels
# ax1.set_title('Northern Hemisphere Sea Ice')
# ax3.set_title('Southern Hemisphere Sea Ice')
ax3.set_xlabel('Age (ka)')
ax1.set_ylabel('Temperature (°C)', color='r')
ax2.set_ylabel('Sea ice area (million km²)', color='b')
ax3.set_ylabel('Temperature (°C)', color='r')
ax4.set_ylabel('Sea ice area (million km²)', color='b')

# Legend
ax1.legend(loc='upper left',fontsize=8)
ax2.legend(loc='upper right',fontsize=8)
ax3.legend(loc='upper left',fontsize=8)
ax4.legend(loc='upper right',fontsize=8)

# set the x-axis limits for both plots
ax1.set_xlim(200,0)
ax3.set_xlim(200,0)
# add verticle gridlines
ax1.grid(axis='x', linestyle='--', linewidth=0.5)
ax3.grid(axis='x', linestyle='--', linewidth=0.5)

# set the linewidth of the box
for axis in ['top','bottom','left','right']:
    ax1.spines[axis].set_linewidth(2)
    ax2.spines[axis].set_linewidth(2)
    ax3.spines[axis].set_linewidth(2)
    ax4.spines[axis].set_linewidth(2)

# add text a, and b to the plot
ax1.text(-0.05, 1.1, 'a', transform=ax1.transAxes, fontsize=16, fontweight='bold', va='top')
ax3.text(-0.05, 1.1, 'b', transform=ax3.transAxes, fontsize=16, fontweight='bold', va='top')

plt.tight_layout()
plt.show()
```
